```python
from sklearn.linear_model import ElasticNet, Lasso,  BayesianRidge, LassoLarsIC
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
n_folds = 5
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

```python
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=1))
```

```python
ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9, random_state=3))
```

```python
import pickle
import math        
from sklearn.model_selection import GridSearchCV
from sklearn.gaussian_process.kernels import WhiteKernel, ExpSineSquared

numStages = 4
ypred = np.zeros((train.shape[0], numStages))

def mm_model_fit():
    T = 2
    for t in range(T):
        print("Start: Iteration %d \n" % (t))
        
        for stage in range(numStages):
            print("Stage %d \n" % (stage))
            target_stage = y_train - ypred[:, np.arange(numStages)!=stage].sum(axis=1)
            print('y_train')
            print(y_train)
            
            print('target_stage')
            print(target_stage)
            
            
            if stage == 0:
                
                best_est = KernelRidge(alpha=1, kernel='polynomial', degree=2, coef0=1)
                
                best_est.fit(train.values,target_stage)
                ypred[:,stage] = best_est.predict(train.values)
                print(ypred)
                
                pickle.dump(best_est, open("stage_"+str(stage), 'wb'))
                print("first step score (RMSE): {}".format(rmsle(y_train,ypred[:,stage])))
            
            elif stage == 1:
                
                best_est = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
                
                best_est.fit(train.values,target_stage)
                ypred[:,stage] = best_est.predict(train.values)
                print(ypred)
                
                pickle.dump(best_est, open("stage_"+str(stage), 'wb'))
                print("first step score (RMSE): {}".format(rmsle(y_train,ypred[:,stage])))
            
            elif stage == 2:
                
                best_est = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
                
                best_est.fit(train.values,target_stage)
                ypred[:,stage] = best_est.predict(train.values)
                print(ypred)
                
                pickle.dump(best_est, open("stage_"+str(stage), 'wb'))
                print("first step score (RMSE): {}".format(rmsle(y_train,ypred[:,stage])))
            
            else:
                if T == 1:
                    break
                
                best_est = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
                
                best_est.fit(train.values,target_stage)
                ypred[:,stage] = best_est.predict(train.values)
                print(ypred)
                print("first step score (RMSE): {}".format(rmsle(y_train,ypred[:,stage])))
                
                
                pickle.dump(best_est, open("stage_"+str(stage), 'wb'))
                
    print("final score (RMSE): {}".format(rmsle(y_train,ypred.sum(axis=1))))
```

```python
ypred = np.zeros((test.shape[0], numStages))
for stage in range(numStages):
    
    if stage == 0:
        loaded_model = pickle.load(open("stage_"+str(stage), 'rb'))
        ypred[:,stage] = loaded_model.predict(test.values)
    elif stage == 1:
        loaded_model = pickle.load(open("stage_"+str(stage), 'rb'))
        ypred[:,stage] = loaded_model.predict(test.values)
    elif stage == 2:
        loaded_model = pickle.load(open("stage_"+str(stage), 'rb'))
        ypred[:,stage] = loaded_model.predict(test.values)
    else:
        loaded_model = pickle.load(open("stage_"+str(stage), 'rb'))
        ypred[:,stage] = loaded_model.predict(test.values)
    
    
final_ypred = ypred.sum(axis=1)
```
